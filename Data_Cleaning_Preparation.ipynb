PRODUCTS

```python
#reading the csv file data using path
path=r"C:\Users\prana\Documents\DataSpark DataSets\Products.csv"
df_products=pd.read_csv(path)
df_products
```

```python
#checking data types
df_products.info()
```

```python
#checking missing values
df_products.isnull().sum()
```

```python
df_products["Unit Cost USD"]=df_products["Unit Cost USD"].astype(str).str.replace("$", "").str.replace(",", "")
df_products["Unit Cost USD"]=df_products["Unit Cost USD"].astype(float)

df_products["Unit Price USD"]=df_products["Unit Price USD"].astype(str).str.replace("$", "").str.replace(",", "")
df_products["Unit Price USD"]=df_products["Unit Price USD"].astype(float)
```

```python
#checking data types
df_products.info()
```

```python
#checking null values
df_products.isnull().sum()
```

```python
#droping the column
df_products.drop(["SubcategoryKey"],axis=1,inplace=True)
```

```python
#verifying whether there are still any null values 
df_products.isnull().sum()
```
